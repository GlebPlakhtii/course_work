```python
def highlight(s):
    
    max = s.max()
    min = s.min()
    
    return ['background-color: green' if v == max else 'background-color: red' if v == min else '' for v in s]

```

```python
df.style.apply(highlight,subset = ['year','suicides_no','gdp_per_capita','suicides/100k pop'])
```

```python
# get rows with min > suicides/100k pop > max
def get_suicides(df,min,max=float('inf')):
    return df[lambda r: (r['suicides/100k pop']<=max) & (r['suicides/100k pop']>=min)]
```

```python
full_df['avarage_age'] = full_df['age'].map(lambda a: (sum(map(int,a.split()[0].split('-')))/2 if a!='75+ years' else 75))

plt.figure(figsize=(15,10))
plt.xlabel("population")
plt.ylabel("suicides/100k pop")

colormap = ['violet','blueviolet','blue','mediumblue','navy','black']
accordance = {9.5 : 0 , 19.5 : 1, 29.5 : 2, 44.5: 3, 64.5 : 4, 75 : 5}
for age in accordance.keys():
    plt.scatter(full_df['population'][full_df['avarage_age']==age],
                full_df['suicides/100k pop'][full_df['avarage_age']==age], 
                alpha = 0.05, c = colormap[accordance[age]], label ='age : '+str(age),s=10)
plt.legend(prop={'size': 15})

plt
```
